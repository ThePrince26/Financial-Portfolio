## Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Code

```python
start_date = "2000-09-01"
end_date = "2023-03-21"
companies = ["CVX", "XOM", "BP", "GE", "NEE"]
company = ["t"]
Data = yf.download(companies, start=start_date, end=end_date)
```

```python
Data
```

## Cleaning Data

```python
Data.dropna()
Data.duplicated()
```

## Selecting Data

```python
Data = Data.Close
```

```python
Data.head()
```

```python
# Use pct_change() to calculate the percentage change of the value over time
pct_change = Data["BP"]

# Select the first and last rows of the resulting dataframe
first_value = pct_change.iloc[1]
last_value = pct_change.iloc[-1]

print (first_value)
print (last_value)

# Calculate the percentage change between the first and last values
percentage_change = (last_value - first_value) / first_value * 100

# Print the result
print(f"The percentage change from the first date to the last date is: {percentage_change:.2f}%")
```
